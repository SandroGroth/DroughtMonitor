## 02. Setting up working directory

```
import os

base_path = "/content/drive/My Drive/DroughtMonitor"

data_path = os.path.join(base_path, "Data")
raster_data_path = os.path.join(data_path, "Raster")
vector_data_path = os.path.join(data_path, "Vector")
normal_NDVI_path = os.path.join(raster_data_path, "Normal_NDVIs")

try:
  os.mkdir(base_path)
  print("Created working directory: %s" % base_path)
  os.mkdir(data_path)
  print("Created Data directory: %s" % data_path)
  os.mkdir(raster_data_path)
  print("Created Data directory: %s" % raster_data_path)
  os.mkdir(vector_data_path)
  print("Created Data directory: %s" % vector_data_path)
  os.mkdir(normal_NDVI_path)
  print("Created Data directory: %s" % normal_NDVI_path)
except OSError as e:
  print("Directory creation failed: %s" % e)
else:
  print("All paths created successfully.")
```
